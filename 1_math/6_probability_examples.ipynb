# 6. Probability in Python

Probability tells us how likely something is to happen!
It's a number between 0 (impossible) and 1 (certain).
Think of it as a way to measure uncertainty!


```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## 1. What is Probability?

Probability = number of ways something can happen / total number of possibilities
Probability is always between 0 and 1:
- 0 = impossible (will never happen)
- 1 = certain (will always happen)
- 0.5 = 50% chance (equally likely to happen or not)


```python
# Example: Flipping a coin
# Two possible outcomes: Heads or Tails
# Each has probability 0.5 (50%)

print("Flipping a coin:")
print("Possible outcomes: Heads, Tails")
print("Probability of Heads = 1/2 = 0.5 = 50%")
print("Probability of Tails = 1/2 = 0.5 = 50%")
print()

# Example: Rolling a die
# Six possible outcomes: 1, 2, 3, 4, 5, 6
# Each has probability 1/6

print("Rolling a die:")
print("Possible outcomes: 1, 2, 3, 4, 5, 6")
print("Probability of rolling 1 = 1/6 ≈ 0.167 = 16.7%")
print("Probability of rolling 2 = 1/6 ≈ 0.167 = 16.7%")
print("Probability of rolling 3 = 1/6 ≈ 0.167 = 16.7%")
print("Probability of rolling 4 = 1/6 ≈ 0.167 = 16.7%")
print("Probability of rolling 5 = 1/6 ≈ 0.167 = 16.7%")
print("Probability of rolling 6 = 1/6 ≈ 0.167 = 16.7%")
print()

# Visualize
outcomes = ['Heads', 'Tails']
probabilities = [0.5, 0.5]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Coin flip
ax1.bar(outcomes, probabilities, color=['gold', 'silver'])
ax1.set_ylim(0, 1)
ax1.set_ylabel('Probability')
ax1.set_title('Flipping a Coin')
ax1.set_yticks([0, 0.25, 0.5, 0.75, 1.0])
ax1.grid(True, alpha=0.3)

# Die roll
die_outcomes = [1, 2, 3, 4, 5, 6]
die_probs = [1/6] * 6
ax2.bar(die_outcomes, die_probs, color='steelblue')
ax2.set_ylim(0, 1)
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Probability')
ax2.set_title('Rolling a Die')
ax2.set_yticks([0, 1/6, 0.5, 1.0])
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 2. Calculating Probability

Probability = favorable outcomes / total outcomes
Let's calculate probabilities for different events!


```python
# Example: Rolling a die
# What's the probability of rolling an even number?

total_outcomes = 6  # 1, 2, 3, 4, 5, 6
favorable_outcomes = 3  # 2, 4, 6 (even numbers)

probability = favorable_outcomes / total_outcomes

print("Rolling a die:")
print(f"Total outcomes: {total_outcomes}")
print(f"Favorable outcomes (even numbers): {favorable_outcomes}")
print(f"Probability = {favorable_outcomes}/{total_outcomes} = {probability} = {probability*100}%")
print()

# Example: Drawing a card from a deck
# Standard deck: 52 cards, 4 suits (Hearts, Diamonds, Clubs, Spades)
# 13 cards per suit (Ace, 2-10, Jack, Queen, King)

print("Drawing a card from a deck:")
print(f"Total cards: 52")

# Probability of drawing a heart
hearts = 13
prob_heart = hearts / 52
print(f"Hearts: {hearts}")
print(f"Probability of Heart = {hearts}/52 = {prob_heart:.3f} = {prob_heart*100:.1f}%")
print()

# Probability of drawing an Ace
aces = 4
prob_ace = aces / 52
print(f"Aces: {aces}")
print(f"Probability of Ace = {aces}/52 = {prob_ace:.3f} = {prob_ace*100:.1f}%")
print()

# Probability of drawing a red card
red_cards = 26  # Hearts + Diamonds
prob_red = red_cards / 52
print(f"Red cards: {red_cards}")
print(f"Probability of Red = {red_cards}/52 = {prob_red:.3f} = {prob_red*100:.1f}%")
```

```python
# Simulate coin flips
num_flips = 1000
results = [random.choice(['Heads', 'Tails']) for _ in range(num_flips)]

heads_count = results.count('Heads')
tails_count = results.count('Tails')

prob_heads = heads_count / num_flips
prob_tails = tails_count / num_flips

print(f"Simulating {num_flips} coin flips:")
print(f"Heads: {heads_count} times")
print(f"Tails: {tails_count} times")
print()
print(f"Probability of Heads ≈ {prob_heads:.3f} = {prob_heads*100:.1f}%")
print(f"Probability of Tails ≈ {prob_tails:.3f} = {prob_tails*100:.1f}%")
print(f"(Theoretical probability = 0.5 = 50%)")
print()

# Simulate die rolls
num_rolls = 1000
die_rolls = [random.randint(1, 6) for _ in range(num_rolls)]

# Count each outcome
outcome_counts = {i: die_rolls.count(i) for i in range(1, 7)}

print(f"Simulating {num_rolls} die rolls:")
for outcome, count in outcome_counts.items():
    prob = count / num_rolls
    print(f"Rolled {outcome}: {count} times, Probability ≈ {prob:.3f} = {prob*100:.1f}%")
print(f"(Theoretical probability = 1/6 ≈ 0.167 = 16.7%)")

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Coin flips
coin_data = [heads_count, tails_count]
ax1.bar(['Heads', 'Tails'], coin_data, color=['gold', 'silver'])
ax1.set_ylabel('Count')
ax1.set_title(f'Coin Flips ({num_flips} flips)')
ax1.grid(True, alpha=0.3)
for i, v in enumerate(coin_data):
    ax1.text(i, v + 10, f'{v}\n({coin_data[i]/num_flips*100:.1f}%)', 
             ha='center', va='bottom')

# Die rolls
outcomes = list(outcome_counts.keys())
counts = list(outcome_counts.values())
ax2.bar(outcomes, counts, color='steelblue')
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Count')
ax2.set_title(f'Die Rolls ({num_rolls} rolls)')
ax2.grid(True, alpha=0.3)
for i, v in enumerate(counts):
    ax2.text(outcomes[i], v + 5, f'{v}\n({v/num_rolls*100:.1f}%)', 
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## 4. Independent Events

Independent events don't affect each other!
The probability of both happening = product of their probabilities.
P(A and B) = P(A) × P(B)


```python
# Example: Flipping two coins
# Probability of getting Heads on first coin = 0.5
# Probability of getting Heads on second coin = 0.5
# Probability of getting Heads on both = 0.5 × 0.5 = 0.25

prob_heads = 0.5
prob_both_heads = prob_heads * prob_heads

print("Flipping two coins:")
print(f"Probability of Heads on first coin: {prob_heads}")
print(f"Probability of Heads on second coin: {prob_heads}")
print(f"Probability of both Heads: {prob_heads} × {prob_heads} = {prob_both_heads}")
print()

# All possible outcomes
outcomes = [
    ('Heads', 'Heads'),
    ('Heads', 'Tails'),
    ('Tails', 'Heads'),
    ('Tails', 'Tails'),
]

print("All possible outcomes:")
for i, outcome in enumerate(outcomes, 1):
    print(f"{i}. {outcome[0]} and {outcome[1]}")

print(f"\nTotal outcomes: {len(outcomes)}")
print(f"Favorable (both Heads): 1")
print(f"Probability = 1/4 = 0.25 = {prob_both_heads*100}%")
print()

# Example: Rolling two dice
# Probability of rolling 1 on first die = 1/6
# Probability of rolling 1 on second die = 1/6
# Probability of both being 1 = (1/6) × (1/6) = 1/36

prob_one = 1/6
prob_both_one = prob_one * prob_one

print("Rolling two dice:")
print(f"Probability of 1 on first die: {prob_one:.3f}")
print(f"Probability of 1 on second die: {prob_one:.3f}")
print(f"Probability of both being 1: {prob_one:.3f} × {prob_one:.3f} = {prob_both_one:.4f} = 1/36")
print()

# Simulate
num_experiments = 10000
both_heads_count = 0
both_one_count = 0

for _ in range(num_experiments):
    # Two coin flips
    coin1 = random.choice(['Heads', 'Tails'])
    coin2 = random.choice(['Heads', 'Tails'])
    if coin1 == 'Heads' and coin2 == 'Heads':
        both_heads_count += 1
    
    # Two die rolls
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 == 1 and die2 == 1:
        both_one_count += 1

print(f"Simulation ({num_experiments} experiments):")
print(f"Both Heads: {both_heads_count} times, Probability ≈ {both_heads_count/num_experiments:.3f}")
print(f"Both 1s: {both_one_count} times, Probability ≈ {both_one_count/num_experiments:.4f}")
print(f"(Theoretical: both Heads = 0.25, both 1s ≈ 0.0278)")
```

## 5. Mutually Exclusive Events

Mutually exclusive events can't happen at the same time!
The probability of either happening = sum of their probabilities.
P(A or B) = P(A) + P(B)


```python
# Example: Rolling a die
# Probability of rolling 1 = 1/6
# Probability of rolling 2 = 1/6
# Probability of rolling 1 OR 2 = 1/6 + 1/6 = 2/6 = 1/3

prob_1 = 1/6
prob_2 = 1/6
prob_1_or_2 = prob_1 + prob_2

print("Rolling a die:")
print(f"Probability of rolling 1: {prob_1:.3f}")
print(f"Probability of rolling 2: {prob_2:.3f}")
print(f"Probability of rolling 1 OR 2: {prob_1:.3f} + {prob_2:.3f} = {prob_1_or_2:.3f} = {prob_1_or_2*100:.1f}%")
print()

# Example: Rolling an even number
# Even numbers: 2, 4, 6
# Each has probability 1/6
# Probability of even = 1/6 + 1/6 + 1/6 = 3/6 = 1/2

prob_even = 3 * (1/6)
print("Probability of rolling an even number (2, 4, or 6):")
print(f"Probability = 1/6 + 1/6 + 1/6 = {prob_even:.3f} = {prob_even*100:.1f}%")
print()

# Example: Drawing a card
# Probability of Heart = 13/52
# Probability of Ace = 4/52
# But Heart OR Ace: need to be careful (Ace of Hearts is in both!)
# Actually, these are NOT mutually exclusive!

prob_heart = 13/52
prob_ace = 4/52
# Heart OR Ace = Hearts + Aces - Ace of Hearts (counted twice)
prob_heart_or_ace = prob_heart + prob_ace - (1/52)  # Subtract overlap

print("Drawing a card:")
print(f"Probability of Heart: {prob_heart:.3f}")
print(f"Probability of Ace: {prob_ace:.3f}")
print(f"Probability of Heart OR Ace: {prob_heart_or_ace:.3f}")
print("(Note: Ace of Hearts is counted in both, so we subtract it once)")
print()

# Mutually exclusive example: Red or Black
prob_red = 26/52
prob_black = 26/52
prob_red_or_black = prob_red + prob_black

print("Mutually exclusive example:")
print(f"Probability of Red: {prob_red:.3f}")
print(f"Probability of Black: {prob_black:.3f}")
print(f"Probability of Red OR Black: {prob_red:.3f} + {prob_black:.3f} = {prob_red_or_black:.3f} = {prob_red_or_black*100:.1f}%")
print("(A card can't be both red and black!)")
```

## 6. Probability Distributions

A probability distribution shows all possible outcomes and their probabilities!
Let's visualize some common distributions!


```python
# Uniform distribution: all outcomes equally likely
# Example: Rolling a die

outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6

print("Uniform Distribution (Rolling a die):")
print("All outcomes have equal probability")
print()

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Uniform distribution
axes[0, 0].bar(outcomes, probabilities, color='steelblue')
axes[0, 0].set_xlabel('Outcome')
axes[0, 0].set_ylabel('Probability')
axes[0, 0].set_title('Uniform Distribution (Die Roll)')
axes[0, 0].set_ylim(0, 0.2)
axes[0, 0].grid(True, alpha=0.3)
for i, p in enumerate(probabilities):
    axes[0, 0].text(outcomes[i], p + 0.01, f'{p:.3f}', ha='center')

# Binomial distribution (simplified: flipping 3 coins)
# Probability of getting k heads in 3 flips
from math import comb

k_values = [0, 1, 2, 3]  # Number of heads
binomial_probs = [comb(3, k) * (0.5**k) * (0.5**(3-k)) for k in k_values]

axes[0, 1].bar(k_values, binomial_probs, color='green')
axes[0, 1].set_xlabel('Number of Heads')
axes[0, 1].set_ylabel('Probability')
axes[0, 1].set_title('Binomial Distribution (3 Coin Flips)')
axes[0, 1].set_xticks(k_values)
axes[0, 1].grid(True, alpha=0.3)
for i, p in enumerate(binomial_probs):
    axes[0, 1].text(k_values[i], p + 0.02, f'{p:.3f}', ha='center')

print("Binomial Distribution (Flipping 3 coins):")
for k, p in zip(k_values, binomial_probs):
    print(f"Probability of {k} heads: {p:.3f} = {p*100:.1f}%")
print()

# Normal distribution (bell curve) - approximated
x = np.linspace(-3, 3, 100)
normal_probs = np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)

axes[1, 0].plot(x, normal_probs, 'r-', linewidth=2)
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Probability Density')
axes[1, 0].set_title('Normal Distribution (Bell Curve)')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].fill_between(x, normal_probs, alpha=0.3, color='red')

# Sum of two dice
sum_outcomes = list(range(2, 13))
sum_probs = []
for s in sum_outcomes:
    # Ways to get sum s with two dice
    ways = 0
    for d1 in range(1, 7):
        for d2 in range(1, 7):
            if d1 + d2 == s:
                ways += 1
    sum_probs.append(ways / 36)

axes[1, 1].bar(sum_outcomes, sum_probs, color='orange')
axes[1, 1].set_xlabel('Sum')
axes[1, 1].set_ylabel('Probability')
axes[1, 1].set_title('Sum of Two Dice')
axes[1, 1].set_xticks(sum_outcomes)
axes[1, 1].grid(True, alpha=0.3)
for i, p in enumerate(sum_probs):
    axes[1, 1].text(sum_outcomes[i], p + 0.01, f'{p:.3f}', ha='center', fontsize=8)

plt.tight_layout()
plt.show()

print("Sum of Two Dice:")
for s, p in zip(sum_outcomes, sum_probs):
    print(f"Sum = {s:2d}: Probability = {p:.3f} = {p*100:.1f}%")
```

## 7. Expected Value

Expected value = average outcome if you repeat an experiment many times!
It's the sum of: (outcome × probability of that outcome)


```python
# Example: Rolling a die
# Expected value = average of all outcomes

outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6

expected_value = sum(outcome * prob for outcome, prob in zip(outcomes, probabilities))

print("Rolling a die:")
print("Outcomes and probabilities:")
for outcome, prob in zip(outcomes, probabilities):
    print(f"  {outcome}: probability = {prob:.3f}")
print()
print(f"Expected value = ", end="")
for i, (outcome, prob) in enumerate(zip(outcomes, probabilities)):
    if i > 0:
        print(" + ", end="")
    print(f"{outcome}×{prob:.3f}", end="")
print(f" = {expected_value:.2f}")
print()
print(f"Expected value = {expected_value:.2f}")
print("This means: if you roll a die many times, the average will be about 3.5")
print()

# Simulate to verify
num_rolls = 10000
rolls = [random.randint(1, 6) for _ in range(num_rolls)]
average = sum(rolls) / len(rolls)

print(f"Simulation ({num_rolls} rolls):")
print(f"Average = {average:.2f}")
print(f"(Expected value = {expected_value:.2f})")
print()

# Example: Flipping coins for money
# You win $1 for Heads, lose $1 for Tails
# Expected value = ?

win_amount = 1
lose_amount = -1
prob_win = 0.5
prob_lose = 0.5

expected_value_game = win_amount * prob_win + lose_amount * prob_lose

print("Coin flip game:")
print("Heads: win $1 (probability = 0.5)")
print("Tails: lose $1 (probability = 0.5)")
print(f"Expected value = ${win_amount} × {prob_win} + ${lose_amount} × {prob_lose}")
print(f"Expected value = ${expected_value_game:.2f}")
print("This means: on average, you break even (fair game)")
```

## 8. Conditional Probability

Conditional probability = probability of A given that B happened!
P(A | B) = probability of A, given B


```python
# Example: Drawing cards
# What's the probability of drawing an Ace, given that we drew a Heart?

print("Drawing a card from a deck:")
print("Total cards: 52")
print()

# Probability of Heart
prob_heart = 13/52
print(f"Probability of Heart: {prob_heart:.3f}")
print()

# Probability of Ace given Heart
# If we know it's a Heart, there are 13 hearts
# One of them is an Ace (Ace of Hearts)
prob_ace_given_heart = 1/13

print("Conditional probability:")
print(f"Probability of Ace, given Heart: {prob_ace_given_heart:.3f}")
print("(Out of 13 hearts, 1 is an Ace)")
print()

# Compare with unconditional probability
prob_ace = 4/52
print(f"Probability of Ace (unconditional): {prob_ace:.3f}")
print(f"Probability of Ace, given Heart: {prob_ace_given_heart:.3f}")
print("(Knowing it's a Heart changes the probability!)")
print()

# Example: Rolling two dice
# What's the probability that sum is 7, given that first die is 4?

print("Rolling two dice:")
print("Probability that sum = 7, given first die = 4:")
print()

# If first die is 4, second die must be 3 for sum to be 7
# Second die has 6 possible outcomes: 1, 2, 3, 4, 5, 6
# One of them gives sum = 7

prob_sum7_given_first4 = 1/6

print(f"Given first die = 4, second die must be 3")
print(f"Probability = 1/6 = {prob_sum7_given_first4:.3f}")
print()

# Compare with unconditional
# Total ways to get sum 7: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1) = 6 ways
# Total outcomes: 36
prob_sum7 = 6/36
print(f"Probability of sum = 7 (unconditional): {prob_sum7:.3f}")
print(f"Probability of sum = 7, given first die = 4: {prob_sum7_given_first4:.3f}")
```

## 9. Key Takeaways

**What probability is:**
- Measure of how likely something is to happen
- Always between 0 and 1
- 0 = impossible, 1 = certain, 0.5 = 50% chance

**Key formulas:**
- **Probability**: P(event) = favorable outcomes / total outcomes
- **Independent events**: P(A and B) = P(A) × P(B)
- **Mutually exclusive**: P(A or B) = P(A) + P(B)
- **Expected value**: E = Σ (outcome × probability)
- **Conditional probability**: P(A | B) = probability of A given B

**Important concepts:**
- **Independent events**: Don't affect each other
- **Mutually exclusive**: Can't happen at the same time
- **Probability distribution**: Shows all outcomes and probabilities
- **Expected value**: Average outcome over many trials
- **Conditional probability**: Probability given some condition

**Remember:** Probability helps us understand uncertainty and make predictions!
